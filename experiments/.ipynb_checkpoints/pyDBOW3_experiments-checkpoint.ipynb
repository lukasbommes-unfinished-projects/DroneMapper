```python
import cv2
import numpy as np
import pyDBoW3 as bow
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("pydbow3_test.png", cv2.IMREAD_COLOR)
```

```python
# extract features
orb = cv2.ORB_create()
fast = cv2.FastFeatureDetector_create(threshold=12)

kp = orb.detect(img, None)
kp, des = orb.compute(img, kp)
```

```python
img_vis = cv2.drawKeypoints(np.copy(img), kp, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
```

```python
f, ax = plt.subplots(figsize=(15, 10))
ax.imshow(img_vis[:, :, ::-1])
plt.show()
```
