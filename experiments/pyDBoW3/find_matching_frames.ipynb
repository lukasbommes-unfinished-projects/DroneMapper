```python
import pyDBoW3 as bow
import cv2
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.append("../..")
from ssc import ssc
```

```python
frames = []
files = ["frame_300.png", "frame_1750.png", "frame_2025.png", "frame_3000.png", "frame_3125.png", "frame_4000.png"]
for f in files:
    frames.append(cv2.imread(f, cv2.IMREAD_COLOR))
```

```python
for i in range(len(frames)):
    frames[i] = cv2.cvtColor(frames[i], cv2.COLOR_BGR2GRAY)
```

```python
voc = bow.Vocabulary()
voc.load("../../pyDBoW3/ORBvoc.txt")
```

```python
db = bow.Database()
db.setVocabulary(voc, False, 0)
```

```python
# extract features
orb = cv2.ORB_create()
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
fast = cv2.FastFeatureDetector_create(threshold=12)
num_ret_points = 3000
tolerance = 0.1
num_matches = 3000

def extract_kp_des(current_frame, fast, orb):
    kp = fast.detect(current_frame, None)
    kp = sorted(kp, key = lambda x:x.response, reverse=True)
    kp = ssc(kp, num_ret_points, tolerance, current_frame.shape[1], current_frame.shape[0])
    kp, des = orb.compute(current_frame, kp)
    return kp, des

des = []
kps = []
for frame in frames:
    kp, d = extract_kp_des(frame, fast, orb)
    kps.append(kp)
    des.append(d)
```

```python
for d in des:
    db.add(d)
```

```python
# query 
query_idx = 1

results = db.query(des[query_idx], -1, -1)
results
```
